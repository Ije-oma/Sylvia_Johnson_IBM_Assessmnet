```python
# C) Come up with interesting analysis to draw comparisons between 2 airlines and help
    # your client win the deal.
# Purpose of analysis: To to draw comparisons between 2 airlines in order to highlight KPIs for client;
# using the operational statistics of the monthly civil aviation survey data with a focus on Total services.

# Strategy: use the year which has the greatest number of hours flown for both airlines, compare the ratio 
# of the total Passenger-kilometres of that year, with the total turbo fuel consumption of that year
# for both airlines.

# airline AC analysis:
# Strategy: filter df_filt_AC by operational statistics being = Passenger-kilometres and year = 2015 
stats = (df_filt_AC['Operational statistics'] == 'Passenger-kilometres')
year = (df_filt_AC['YEAR'] == '15')
air_AC= df_filt_AC.loc[stats & year]
# find the total passenger kilometer for 2015
air_AC_sum = air_AC ['VALUE'].sum()
print('The total passenger kilometer for chosen airline in 2015 is {0}'.format (int(air_AC_sum)))
# find the ratio of the total passenger kilometer to total turbo fuel consumption calculated in the 
# previous step; air_AC_sum : TFC_2015
a = air_AC_sum 
b = TFC_2015
def ratio(a, b):
    a = float(a)
    b = float(b)
    if b == 0:
        return a
    return ratio(b, a % b)
print('Ratio of total passenger kilometer to total turbo fuel consumption for chosen airline is {0}'.format (ratio(b, a % b)))

# airline WJ analysis:
# Strategy: filter df_filt_WJ by operational statistics being = Passenger-kilometres and year = 2016
stats_1 = (df_filt_WJ['Operational statistics'] == 'Passenger-kilometres')
year_1 = (df_filt_WJ['YEAR'] == '16')
air_WJ= df_filt_WJ.loc[stats_1 & year_1]
# find the sum passenger kilometer for 2016
air_WJ_sum = air_WJ ['VALUE'].sum()
print('The total passenger kilometer for competitor airline in 2016 is {0}'.format (int(air_WJ_sum)))
# find the ratio of the total passenger kilometer to total turbo fuel consumption calculated in the 
# previous step; air_WJ_sum : TFC_2016
c = air_WJ_sum
d = TFC_2016
def ratio(c, d):
    c = float(c)
    d = float(d)
    if d == 0:
        return c
    return ratio(d, c % d)
print('Ratio of total passenger kilometer to total turbo fuel consumption for competior airline is {0}'.format (ratio(d, c % d)))
```
